## Ejercicio 2
Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.




```
import pandas as pd
alumnos = ['Mario','Bogdan','Mateo','Fernanda']
notas = {}
for i in alumnos:
    notas[i] = float(input(f"Calificación de {i}: "))
notas = pd.Series(notas)
notas = pd.Series([notas.min(),notas.max(),notas.mean(),notas.std()],index=['Mínimo','Máximo','Media','Desviación típica'])
print(notas)
```
