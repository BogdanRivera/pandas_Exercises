## Ejercicio 5
Escribir una funci√≥n que reciba un DataFrame con el formato del ejercicio anterior, una lista de meses, y devuelva el balance (ventas - gastos) total en los meses indicados.

```
def balance(data,meses):
    data['Balance'] = data.Ventas - data.Gastos
    return data.set_index('Mes').loc[meses,'Balance'].sum()
    #return data[data.Mes.isin(meses).Balance.sum()] 



import pandas as pd
data = {'Mes':['Enero','Febrero','Marzo','Abril'],
'Ventas':[30500,35600,28300,33900],
'Gastos':[22000,23400,18100,20700]
}
df = pd.DataFrame(data)
meses = ['Enero','Marzo']
print(balance(df,meses))
```
