## Ejercicio 3
Escribir una funciÃ³n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con las notas de los alumnos aprobados ordenadas de mayor a menor.

```
import pandas as pd
```

```
def is_aprobated(alumnos):
    alumnos = pd.Series(alumnos)
    alumnos = alumnos[alumnos>6] 
    return alumnos.sort_values(ascending=False)

if __name__=="__main__":
    calificaciones = {'Bogdan':5.4,'Fernanda':10.0,'Diego':7.2,'Delia': 9.3,'Alejandro':5.5} 
    print(is_aprobated(calificaciones))
```
